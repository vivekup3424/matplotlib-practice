```python
import numpy as np
import matplotlib.pyplot as plt

# Given constants
cin = 3
V = 28
F = 4*12
C0 = 10
tspan = [0, 4]
N = 100

# Time vector
t = np.linspace(tspan[0], tspan[1], N)

# Differential equation function
def lake_pollution(t, c):
    return (F * cin / V) - (F * c / V)

# Initialize concentration array
c = np.zeros(N)
c[0] = C0  # Initial condition

# Time step
dt = (tspan[1] - tspan[0]) / (N - 1)

# Euler's method to solve the differential equation
for i in range(1, N):
    c[i] = c[i-1] + lake_pollution(t[i-1], c[i-1]) * dt

# Plotting the results
plt.plot(t, c, label="Concentration of pollutants")
plt.xlabel("Time (t)")
plt.ylabel("Concentration (c)")
plt.title("Simulation of Lake Pollution")
plt.legend()
plt.grid(True)
plt.show()




```

Here's the code using `solve_ivp` to solve the lake pollution problem:

### Explanation:
1. **Differential Equation**: The rate of change of concentration, \( \frac{dc}{dt} = \frac{F \cdot cin}{V} - \frac{F \cdot c}{V} \), is defined in the `lake_pollution` function.
2. **Initial Condition**: \( C0 = 10 \), the initial concentration of pollutants in the lake.
3. **Time Span**: From `t = 0` to `t = 4` (days or hours, depending on context).
4. **`solve_ivp`**: This function solves the differential equation. The `t_eval` argument defines the times at which to evaluate the solution.
5. **Plotting**: The result is plotted showing how the concentration of pollutants in the lake changes over time.

This code will give a more accurate solution to the lake pollution problem using `solve_ivp`, compared to the Euler method.

```python
### Code for solving the lake pollution problem using `solve_ivp`:

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Given constants
cin = 7
V = 28
F = 4 * 12  # Flow rate
C0 = 10     # Initial concentration
t_span = [0, 40]  # Time span

# Differential equation for lake pollution
def lake_pollution(t, c):
    return (F * cin / V) - (F * c / V)

# Solve the differential equation using solve_ivp
sol = solve_ivp(lake_pollution, t_span, [C0], t_eval=np.linspace(t_span[0], t_span[1], 100))

# Extract solution
t = sol.t
c = sol.y[0]

# Plotting the results
plt.plot(t, c, label="Concentration of pollutants (solve_ivp)")
plt.xlabel("Time (t)")
plt.ylabel("Concentration (c)")
plt.title("Simulation of Lake Pollution")
plt.legend()
plt.grid(True)
plt.show()

```
