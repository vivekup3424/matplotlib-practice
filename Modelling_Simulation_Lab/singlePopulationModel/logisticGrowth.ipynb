```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def logistic_growth(t, P, r, K):
    # Logistic growth model
    dPdt = r * P * (1 - P / K)
    return dPdt

# Initial conditions
P0 = 1000  # Initial population
r = 0.5    # Growth rate
K = 10000  # Carrying capacity
t_span = (0, 100)  # Time span
step_size = 1

# Solve the ODE using solve_ivp
t_eval = np.arange(t_span[0], t_span[1] + step_size, step_size)
sol = solve_ivp(lambda t, P: logistic_growth(t, P, r, K), t_span=t_span, y0=[P0], t_eval=t_eval)

# Extract solution
t = sol.t
P = sol.y[0]

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(t, P, label=f'Logistic Growth (r={r}, K={K})')
plt.xlabel('Time (years)')
plt.ylabel('Population')
plt.title('Logistic Growth Model')
plt.grid(True)
plt.legend()
plt.show()
```
