```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def model(t, P, r):
    dPdt = r * P
    return dPdt

# Initial conditions and parameters
P0 = 1000
r = 0.5
t_span = (0, 10)  # Reduced time span
step_size = 1

# Solve the ODE using solve_ivp
t_eval = np.arange(t_span[0], t_span[1] + step_size, step_size)
sol = solve_ivp(lambda t, P: model(t, P, r), t_span=t_span, y0=[P0], t_eval=t_eval)

# Extract solution
t = sol.t
P = sol.y[0]

# Plotting the results
fig, ax = plt.subplots()
ax.plot(t, P)
plt.xlabel('Time (years)')
plt.ylabel('Population')
plt.title('Population Growth Simulation')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def model(t, P, r):
    dPdt = r * P
    return dPdt

# Initial conditions
P0 = 1000
step_size = 1

# Different values for r and t_span
r_values = [0.1, 0.3, 0.5, 0.7]
t_spans = [(0, 50), (0, 100), (0, 200), (0, 300)]

# Plot results for each combination of r and t_span
fig, ax = plt.subplots(figsize=(10, 6))

for r in r_values:
    for t_span in t_spans:
        # Solve the ODE using solve_ivp
        t_eval = np.arange(t_span[0], t_span[1] + step_size, step_size)
        sol = solve_ivp(lambda t, P: model(t, P, r), t_span=t_span, y0=[P0], t_eval=t_eval)
        
        # Extract solution
        t = sol.t
        P = sol.y[0]
        
        # Plotting the results
        ax.plot(t, P, label=f'r={r}, t_span={t_span}')
        
# Configure plot
plt.xlabel('Time (years)')
plt.ylabel('Population')
plt.title('Population Growth for Different r and t_span Combinations')
plt.legend()
plt.grid(True)
plt.show()

```
