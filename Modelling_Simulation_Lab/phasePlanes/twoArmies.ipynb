```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Parameters
a1 = 0.0544
a2 = 0.0106

# Differential equations
def rhs(t, u):
    R, B = u
    dR_dt = -a1 * B
    dB_dt = -a2 * R
    return [dR_dt, dB_dt]

# Time span and step size
time_span = [0, 30]
step_size = 0.1

# Initial conditions
initial_conditions = [
    [66, 18],  # Initial condition 1
    [45, 18],  # Initial condition 2
    [30, 18]   # Initial condition 3
]

# Plot the phase plane for each initial condition
fig, ax = plt.subplots()

for u0 in initial_conditions:
    solution = solve_ivp(
        fun=rhs,
        t_span=time_span,
        y0=u0,
        t_eval=np.arange(time_span[0], time_span[1] + step_size, step_size)
    )
    ax.plot(solution.y[0], solution.y[1], label=f'Initial: {u0}')

# Set labels, title, and grid
ax.set(xlabel='R (Army R)', ylabel='B (Army B)', title='Phase Plane of Two-Armies Model')
ax.legend()
ax.grid()

# Show the plot
plt.show()
```
