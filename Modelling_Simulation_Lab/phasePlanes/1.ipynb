```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Differential equations
def linear_system(t, z):
    X, Y = z
    dX_dt = Y
    dY_dt = -X
    return [dX_dt, dY_dt]

# Time span and step size
time_span = [0, 10]
step_size = 0.1

# Initial conditions for plotting trajectories
initial_conditions = [
    [1, 0],    # Initial condition 1
    [0, 1],    # Initial condition 2
    [-1, 0],   # Initial condition 3
    [0, -1]    # Initial condition 4
]

# Solve the differential equations for each initial condition
fig, ax = plt.subplots()

for initial_condition in initial_conditions:
    solution = solve_ivp(
        fun=lambda t, z: linear_system(t, z),
        t_span=time_span,
        y0=initial_condition,
        t_eval=np.arange(time_span[0], time_span[1] + step_size, step_size)
    )
    ax.plot(solution.y[0], solution.y[1], label=f'Initial: {initial_condition}')

# Plot the equilibrium point
ax.plot(0, 0, 'ro', label='Equilibrium Point (0,0)')

# Set labels, title, and grid
ax.set(xlabel='X', ylabel='Y', title='Phase Plane of Linear System')
ax.legend()
ax.grid()

# Show the plot
plt.show()
```
