```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def model(t, P, r, h):
    dPdt = r * P - h
    return dPdt

P0 = 240_000 #suppose this is the base case
r = 0.35
t_span = (0, 100)  # Reduced time span
step_size = 1
h = 2100

# Solve the ODE using solve_ivp
t_eval = np.arange(t_span[0], t_span[1] + step_size, step_size)
sol = solve_ivp(lambda t, P: model(t, P, r,h), t_span=t_span, y0=[P0], t_eval=t_eval)

# Extract solution
t = sol.t
P = sol.y[0]

# Plotting the results
fig, ax = plt.subplots()
ax.plot(t, P)
plt.xlabel('Time (years)')
plt.ylabel('Population')
plt.title('Population Growth Simulation')
plt.grid()
plt.show()

```
